```python
import pandas as pd 
import numpy as np 
import matplotlib as plt
import seaborn as sbn
```

```python
dataset = pd.read_csv('C://Users//wesle//Documents//dados//marketing_campaign.csv', sep = ';')
df = pd.DataFrame(dataset)
df.info()
```

```python
df_clean = df.drop(['ID','AcceptedCmp3','AcceptedCmp5', 'AcceptedCmp4','AcceptedCmp1','AcceptedCmp2', 'AcceptedCmp4','Complain','Z_CostContact','Z_Revenue','Response'], axis = 1)
display(df_clean)
```

```python
#Tratando valores nulos:
df_clean['Year_Birth'].fillna('Unknown', inplace = True)
df_clean['Education'].fillna('Unknown', inplace = True)
df_clean['Marital_Status'].fillna('Unknown', inplace = True)
#Remove os valores nulos:
# df_netflix.dropna(subset=['Kidhome','rating','duration'], inplace=True)
display(df_clean)
```

```python
df_grdt = df_clean.loc[df_clean['Marital_Status'] == 'Together']
# df_grdt.describe()
df_type = df_grdt.Education.value_counts().to_frame()
print(df_type)
```

```python
# Plotagem do valor de pessoas solteiras e com graduação
fig1 = df_type.plot.pie(y ='Education', autopct = '%0.01f%%', shadow=True, figsize=(10,8))
plt.title('Eduication X Marital_Status', size=20)
plt.show(fig1)
```
